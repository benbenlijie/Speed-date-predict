```python
### build network

pe_a = tf.placeholder(tf.float32, shape=[None, len(feature_df.columns)], name="pe_a")
pe_b = tf.placeholder(tf.float32, shape=[None, len(feature_df.columns)], name="pe_b")
dropout = tf.placeholder(tf.float32, name="dropout")

inputs_ab = tf.concat([pe_a, pe_b], 1)
inputs_ba = tf.concat([pe_b, pe_a], 1)
labels = tf.placeholder(tf.float32, shape=[None, 1], name="labels")
```

```python
input_len = inputs_ab.get_shape()[1].value
hidden_1_len = 100
hidden_2_len = 100
output_count = 1
```

```python
def add_layer(x_tensor, weights, biases):
    output = tf.matmul(x_tensor, weights) + biases
    output = tf.nn.relu(output)
    return output
```

```python
weights = {
    "fc1": tf.Variable(tf.truncated_normal([input_len, hidden_1_len])),
    "fc2": tf.Variable(tf.truncated_normal([hidden_1_len, hidden_2_len])),
    "fc3": tf.Variable(tf.truncated_normal([hidden_2_len, output_count])),
}
biases = {
    "b1": tf.Variable(tf.zeros(hidden_1_len)),
    "b2": tf.Variable(tf.zeros(hidden_2_len)),
    "b3": tf.Variable(tf.zeros(output_count)),
}


hidden_1_layer_ab = add_layer(inputs_ab, weights["fc1"], biases["b1"])
hidden_2_layer_ab = add_layer(hidden_1_layer_ab, weights["fc2"], biases["b2"])
hidden_2_layer_ab = tf.nn.dropout(hidden_2_layer_ab, keep_prob=dropout)

hidden_1_layer_ba = add_layer(inputs_ba, weights["fc1"], biases["b1"])
hidden_2_layer_ba = add_layer(hidden_1_layer_ba, weights["fc2"], biases["b2"])
hidden_2_layer_ba = tf.nn.dropout(hidden_2_layer_ba, keep_prob=dropout)


digits = tf.matmul(hidden_2_layer_ab, weights["fc3"]) + tf.matmul(hidden_2_layer_ba, weights["fc3"]) + biases["b3"]


output = tf.nn.sigmoid(digits)
cast_predict = tf.cast(output > 0.5, tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(cast_predict, labels), tf.float32))

global_step = tf.Variable(0, trainable=False)
learing_rate = tf.train.exponential_decay(0.1, global_step, 1000, 0.8, staircase=True)

cost = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=labels, logits=digits))
#optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
optimizer = tf.train.MomentumOptimizer(learing_rate, 0.9).minimize(cost, global_step=global_step)
```

```python
TP = tf.count_nonzero(cast_predict * labels)
FP = tf.count_nonzero(cast_predict * (labels - 1))
FN = tf.count_nonzero((cast_predict - 1) * labels)

precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)
```

```python
def get_profile(iid):
    return feature_df.loc[iid, :].values

def gen_profile(match_pair):
    iids, pids = [], []
    for [iid, pid] in match_pair:
        if (np.isnan(iid) or np.isnan(pid)):
            continue
        iids.append(get_profile(iid))
        pids.append(get_profile(pid))
    return np.array(iids), np.array(pids)


#iids, pids = gen_profile(train_data.loc[0:10, ['iid', 'pid']].values)
```

```python
n_epochs = 500
```

```python
#%debug
#%pdb on
import random
cost_record = []
sess = tf.Session()

sess.run(tf.global_variables_initializer())
for i in range(n_epochs):
    cv_index = random.randint(0, batch_count)
    for j in range(batch_count):
        if j == cv_index:
            continue;
        """print((j+1)*batch_size)"""
        batch_data = train_data.loc[(j*batch_size):((j+1)*batch_size), :]


        iids, pids = gen_profile(train_data.loc[(j*batch_size):((j+1)*batch_size),
                                                ['iid', 'pid']].values)
        match_label = np.array(train_data.loc[(j*batch_size):((j+1)*batch_size), 
                                              'match'].values)
        match_label = np.reshape(match_label, [len(match_label), 1]).astype(np.float)
        """print(match_label.shape, match_label.dtype)
        print(iids.shape, iids.dtype)
        print(pids.shape, pids.dtype)"""
        sess.run(optimizer, feed_dict={
                pe_a:iids,
                pe_b:pids,
                labels:match_label,
                dropout:0.5
            })
    cv_iids, cv_pids = gen_profile(train_data.loc[(cv_index*batch_size):((cv_index+1)*batch_size),
                                                 ['iid', 'pid']].values)
    match_label = np.array(train_data.loc[(cv_index*batch_size):((cv_index+1)*batch_size),
                                          'match'].values)
    match_label = np.reshape(match_label, [len(match_label), 1]).astype(np.float)
    """print(match_label)"""
    cost_result = sess.run(cost, feed_dict={
            pe_a:cv_iids,
            pe_b:cv_pids,
            labels:match_label,
            dropout:1.0
        })
    cost_record.append(np.mean(cost_result))
    f1_result = sess.run(f1, feed_dict={
            pe_a:cv_iids,
            pe_b:cv_pids,
            labels:match_label,
            dropout:1.0
        })
    print("epoch:", i, "cv_index:", cv_index, "cost:", cost_record[-1], "f1 score:", f1_result)

cost_array = [np.mean(cost_record[i]) for i in range(len(cost_record))]        
```
